Apply a ridge regression model

```python
from sklearn.linear_model import Ridge

#lambda->alpha (renamed in python as lambda is an existing keyword)
rr=Ridge(alpha=.1)
```

```python
predictors=weather.columns[~weather.columns.isin(["target","name","station"])]
```

```python
predictors
```

we have time series data (today's temperature is closely linked to tomorrow's temperature)

cross validation for ML models does not work on such data

we need to use back testing (time series cross validation) to account for the time series nature of the data

Training our Machine Learning Model

*training and testing sets*

*model fitting*

```python
def backtest(weather,model,predictors,start=3650,step=90):
  all_predictions=[]

  for i in range(start,weather.shape[0],step):
    train=weather.iloc[:i,:] #training set
    test=weather.iloc[i:(i+step),:] #testing set

    model.fit(train[predictors],train["target"]) #fit the model to the data

    preds=model.predict(test[predictors])

    preds=pd.Series(preds,index=test.index)
    combined=pd.concat([test["target"],preds],axis=1)

    combined.columns=["actual","prediction"]

    combined["diff"]=(combined["prediction"]-combined["actual"]).abs()

    all_predictions.append(combined)
  return pd.concat(all_predictions)
```

Making predictions

```python
predictions=backtest(weather,rr,predictors)
```

^ ridge regression models run pretty quickly

```python
predictions
```

*accuracy*

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(predictions["actual"],predictions["prediction"])
```

```python
predictions["diff"].mean()
```

^ on average, we were 5 degrees off from the correct temperature

Improving accuracy

*adding predictors*

```python
def pct_diff(old, new):
    return (new - old) / old

def compute_rolling(weather, horizon, col): #rolling mean
    label = f"rolling_{horizon}_{col}" #format strings make it easy to create dynamic strings
    weather[label] = weather[col].rolling(horizon).mean()
    weather[f"{label}_pct"] = pct_diff(weather[label], weather[col])
    return weather

rolling_horizons = [3, 14]
for horizon in rolling_horizons:
    for col in ["tmax", "tmin", "prcp"]:
        weather = compute_rolling(weather, horizon, col)
```

```python
weather
```

```python
weather=weather.iloc[14:,:] #index by no.
```

```python
weather
```

```python
weather=weather.fillna(0)
```

```python
def expand_mean(df):
    return df.expanding(1).mean()

for col in ["tmax", "tmin", "prcp"]:
    weather[f"month_avg_{col}"] = weather[col].groupby(weather.index.month, group_keys=False).apply(expand_mean)
    weather[f"day_avg_{col}"] = weather[col].groupby(weather.index.day_of_year, group_keys=False).apply(expand_mean)
```
